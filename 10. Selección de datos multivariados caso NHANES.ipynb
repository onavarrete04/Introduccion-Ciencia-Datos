```python
import numpy as np 
import pandas as pd
pd.set_option("display.max_columns",100) # muestra todas las columnas cuando vemos el data frame
```

```python
df = pd.read_csv("nhanes_2015_2016.csv")
```

```python
df.head()
```

```python
# obtener el nombre de las columnas

col_names = df.columns
col_names
```

```python
# Una forma de obtener los nombres de las columnas que queremos conservar es simplemente copiando de la salida anterior y almacenando en una lista

conservar = ['BMXWT', 'BMXHT', 'BMXBMI', 'BMXLEG', 'BMXARML', 'BMXARMC',
       'BMXWAIST']
```

```python
# Se puede hacer con comprensión de listas, generando un parametro

[columna for columna in col_names if "BMX" in columna]

# Se esta indicando, haca una columna, ciclo for columnas en lista columna name si 
# lista columnas name, tiene elemento "BMX"
```

```python
# Para mantener las columnas podemos usar []

df_BX = df[conservar] # aqui básicamente se esta conservando todos los datos BMX
```

```python
df_BX.head()
```

Existen dos metodos para seleccionar por fila y por columnas

+ df.loc[filas etiquetas o boleano, columnas etiquetas o boleanos] -> cabe decir que esta espera dos parametros al menos
+ df.iloc[filas enteros o boleanos, columnas enteros o boleanos] -> igualmente espera dos parametros

+ .loc principalmente se basa en la etiqueta, pero también puede utilizarse un array boleano


---------------

+ .iloc se basa principalmente en la posición del index como entero, pero igualmente, puede utilizarse como array boleana

```python
# .loc ->

df.loc[:,conservar].head() # se baso en las etiquetas
```

```python
boleano = np.isin(df.columns,conservar)
boleano
```

```python
df.iloc[:,boleano].head() # index es la lista boleana anterior
```

# Seleccionando por condiciones

```python
# Solo vamos a observar las filas de la variable BMXWAIST 
# sean mayor a la mediana.

v_mediana = pd.Series.median(df_BX["BMXWAIST"])
v_mediana # obtenemos la mediana
```

```python
df_BX[df_BX["BMXWAIST"] > v_mediana].head() 
```

```python
condicion1 = df_BX["BMXWAIST"]>v_mediana
# podemos hacer más condiciones o agregar más
condicion2 = df_BX["BMXLEG"] < 32

df_BX[condicion1 & condicion2].head()
# directamente pasamos las condiciones al data frame
# y este nos da la información
```

```python
df_BX.loc[condicion1 & condicion2,:].head()
# En este segundo caso, estamos pasando al data frame el 
# metodo .loc, indicando las filas a conservar
```

```python
# Marco de datos para ver diferencias entre metodo loc y .loc

variable_tmp = df_BX.loc[condicion1 & condicion2,:].head()
variable_tmp.index = ["a","b","c","d","e"]
# si se llegaran a utilizar más datos esto podría dar un error
# esto podría deberse a que el index sea mucho mayor en otros años
variable_tmp
```

```python
# llamando con los index
variable_tmp.loc[["a","b"],"BMXLEG"]
```

Este error se presenta porque no es un entero o un booleano, por ese motivo es un valor inexacto. 

```python
variable_tmp.iloc[:,2]
```

```python
variable_tmp.loc[:, 2]
# 2 no es un nombre de una columna, en este caso hay un KeyError
```

```python
# Esta es una forma de usar booleanos como lista
# de indexación.
variable_tmp.loc[:,[False, False, True] + [False]*4]
```

```python
variable_tmp.iloc[:, 2]
```

```python
# Podemos cambiar los metodos .loc y .iloc para
# cambiar los valores dentro del marco de datos

variable_tmp.iloc[0:3,2] = [0]*3
variable_tmp.iloc[:,2]
```

```python
variable_tmp.loc["a":"c","BMXBMI"] = [1]*3
variable_tmp.loc[:,"BMXBMI"]
```

```python
# Podemos utilizar el método [] cuando cambiamos todos los valores de una columna
variable_tmp["BMXBMI"] = range(0,5)
variable_tmp
```

```python
# Se dará una advertencia al utilizar [] con 
# condiciones para establecer nuevos valores
# en nuestro conjunto de dataframe

variable_tmp[variable_tmp.BMXBMI > 2]["BMXBMI"] = [10]*2 #  Establecer nuevos valores en una copia de tmp, pero no en la propia tmp
variable_tmp
# El código de la variable anterior no cambio nuestro dataframe tmp
# Esta no es la forma correcta en como se hacen las cosas
```

```python
# La forma en la que se debe hacer las cosas es la siguiente

variable_tmp.loc[variable_tmp.BMXBMI > 2, "BMXBMI"] = [10]*2
variable_tmp
```
