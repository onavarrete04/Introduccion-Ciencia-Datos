### Trabajar con Datos

Es sumamente importante poder trabajar con porciones de datos específicos, para lo cuál existen 3 funciones:

1. .loc()
2. .iloc()
3. .ix()


### .loc()

Toma dos operadores de rango de lista única que estan separados por una coma

```python
df.loc[910:,"razon_social_establecimiento"]

# mostrara todas las observaciones de esta fila
# los primeros dos puntos [:] indican el numero de filas
# en este caso son todas, y despues de la coma, indica
# el nombre de la columna.
# si es [:10,"nombre_columna"] -> mostrará los primeros 10
# si es [10:,"nombre_columna"] -> mostrará todos desde el 10
```

```python
# dará error -> df.loc[:,"ano","estado_rnt","razon_social_establecimiento"]
# Muchas indexaciones 

# Forma correcta.

df.loc[:,["ano","estado_rnt","razon_social_establecimiento"]]

df.loc[df.estado_rnt=="ACTIVO"]
```

## .iloc()

Segmentación basadas en enteros, en el ejemplo anterior con .loc() se localizaban las etiquetas, en este caso por medio de un número entero se indica los datos que querremos ver.


```python
df.iloc[:4]
```

```python
df.iloc[1:4, 2:4] 

# las primeras 4 filas, iniciando desde el index 1
# pero las posiciones 2 y 4 en columnas, osea dos columnas
```
