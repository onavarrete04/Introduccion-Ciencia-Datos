# Muestreo aleatorio simple

Para iniciar podemos iniciar con un marco de muestreo / lista. Para ello comenzaremos a trabajar o a tener en cuenta algunas variables de convenciones.

+ **N**= Población.
+ **n**= Muestra.

Cuando hacemos un muestreo aletorio simple, lo que haremos es seleccionar **n** aleatoriamente. 

En este caso, cada diseño de muestra tiene una probabilidad igual de selección para cada unidad en la lista.

> > Probabilidad de cada unidad = **n/N** 

Esto quiere decir que todas las muestras de tamaño n son igualmente probables. 

#### Estimaciones de media, mediana, proporciones, totales.

Estas medidas las denominaremos como medidas imparticiales cuando surjan de una muestra aleatoria, esto quiere decir que estas medidas de la muestra son iguales a los valores de la población. 

Ahora, en este tipo de muestra aleatoria simple las unidades seleccionadas pueden ser remplazadas o no.

**Remplazadas:** Se eligen para realizar la muestra y luego se devuelven a la población para que puedan ser nuevamente seleccionadas. Las muestras a realizarse pueden ser infinitas.

**No remplazadas:** Se eligen de la población, pero al terminar la muestra no vuelven a la población, lo que quiere decir que las muestras a realizarse son finitas.

En este tipo de elección de muestras lo normal es utilizar la opción no remplazadas, en la realidad este tipo de muestreo aleatorio simple no se suele usar, sobre todo cuando existe una población muy grande, debido a que puede ser extremadamente costoso.

# Complejas - Muestreo aleatorio 

Cuando las poblaciones son grandes y se busca que la muestra sea pequeña, puede ser muy costoso realizar utilizar este tipo de metodos de muestreo probabilistico simple, ahora con las muestras "complejas" se realizan o aplican mecanismos para que el proceso pueda ser más barato. 

Cuando se realizan las muestras COMPLEJAS estas en su diseño suelen desviarse de los principios del muestreo aleatorio simple, 

+ Tienen características clave, estas características distinguen los tipos de muestras respecto a las muestras aleatorias simples.

En el timpo de muestras complejas, la muestra tiene una característica de estrato y cada parte de la muestra se asigna por cada estrato, así se reduce la varianza (medida de dispersión) de las estimaciones de la encuesta. Esta técnica es conocida como estratificación. 
+ Unidades de conglomerado.

En este caso podrían usarse por ejemplo unidades de medida como los departamentos, donde haya una probabilidad conocida de selección dentro de cada diferente estrato, y luego se definiria una muestra de hogares o empresas dentro de ese departamento para recopilar información.
