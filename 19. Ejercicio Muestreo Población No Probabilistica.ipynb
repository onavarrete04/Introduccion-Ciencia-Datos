```python
# importamos las librerias

import numpy as np # Para muestreo de las distribuciones
import matplotlib.pyplot as plt # para trazar grficas basicas
import seaborn as sns; sns.set() # trazar histogramas
%matplotlib inline 

#Creando simulacin

media_uni = 155 # representa el peso
sd_uni = 5 # representa la desviacin estandar
media_gym = 185 # peso de personas que van al gym -> mas musculosas
sd_gym = 5 
gymperc = .3 # 30% de personas que visitan el gym
totalPopSize = 40000 #tamao total de poblacion

# Creando los dos subgrupos
uni_estudiantes = np.random.normal(media_uni, sd_uni, int(totalPopSize * (1-gymperc)))
# normal -> significa que va a generar los datso aleatorios
# de una distribucin normal, parametros (media, desviacion estandar), (poblacion total * 30%)
estudt_van_gym = np.random.normal(media_gym,sd_gym, int(totalPopSize * (gymperc))) 

# Crea la poblacin de los subgrupos

poblacion = np.append(uni_estudiantes, estudt_van_gym)

# conjunto para trazar la figura
plt.figure(figsize=(10,12))

# Grfico Solo estudiantes
plt.subplot(3,1,1) # posicin donde se dibujara
sns.histplot(uni_estudiantes,kde=True, stat="density", linewidth=0)
plt.title("Solo estudiantes Uni")
plt.xlim([140,200])

# Grfico estudiantes van al gym
plt.subplot(3,1,2) # posicin donde se dibujara
sns.histplot(estudt_van_gym,kde=True, stat="density", linewidth=0)
plt.title("Estudiantes Uni al GYM")
plt.xlim([140,200])

# Ambos grficos
plt.subplot(3,1,3) # posicin donde se dibujara
sns.histplot(poblacion,kde=True, stat="density", linewidth=0)
plt.title("Poblacin completa de estudiantes")
plt.axvline(x=np.mean(poblacion))
plt.xlim([140,200])

plt.show()
```


# Muestra aleatoria de la poblacin

En este ejercicio se tomar una muestra completa de todos los estudiantes.

```python
# Simulacin de parametros
numeroMuestras = 5000 #veces que se va a repetir la muestra una y otra vez en el ejercicio
tamaoMuestra = 50

## Obtenga la distribucin de muestreo de la media de slo el gimnasio

media_distribucion = np.empty(numeroMuestras) # se esta creando un vector
for i in range(numeroMuestras):
    estudiantes_random = np.random.choice(poblacion,tamaoMuestra)
    media_distribucion[i] = np.mean(estudiantes_random)
#
# Traza la poblacin y la distribucin de muestreo sesgada

plt.figure(figsize=(10,8))

# trazando la poblacin otra vez

plt.subplot(2,1,1)
sns.histplot(poblacion, kde=True, stat="density", linewidth=0)
plt.title("Total poblacin de estudiantes Uni")
plt.axvline(x=np.mean(poblacion)) # linea de la media
plt.xlim([140,200]) # valores del eje x

# trazando la muestra de distribucin 

plt.subplot(2,1,2)
sns.histplot(media_distribucion,kde=True, stat="density", linewidth=0)
plt.title("Muestra de Distribucion de la Media del peso de Todos Estudiantes")
plt.axvline(x=np.mean(poblacion), color="yellow")
plt.axvline(x=np.mean(media_distribucion),color = "red")
plt.xlim([140,200])

plt.show()
```
